# 1. Revisión del material Double Deep Q-Learning

Utilizando el planteamiento original de la IA implementada en la práctica (Deep Q-Learning). Los componentes del entorno y sus valores son los siguientes:

- **Conjunto de Acciones** -> {comprar, vender, esperar}
- **Inventario** -> Acciones en el portafolio
- **Beneficios (Objetivo, profits)** -> Deferencia entre valor comprado y vendido
- **Reinicio (Reset)** -> Reinicia el entorno y devuelve el estado inicial
- **Estado del entorno** -> Un vector real de once elementos
- **Estado terminal** -> Último registro

Debemos tomar en cuenta que en la medida en que nuestro espacio de estados creo en tamaño, el tiempo de atravesar todos esos estados y la actualización iterativa de todos los valores Q (o valores de la acción) también crecerán en número, como elementos cruciales que representan el premio acumulado esperado que el agente inteligente puede obtener por tomar una acción particular en un estado dado y siguiendo una **política** específica (que define el comportamiento del agente inteligente en un ambiente específico, indicando las acciones que deben ser tomadas por el agente cuando se encuentra en un estado dado).

Cuando tenemos un ambiente complejo, en lugar de utilizar una función de aproximación para estimar la **función Q(s,a)** donde **s** representa un estado y **a** representa una acción; y que provee una estimación del valor de tomar una acción específica en un estado dado.

La Q(s,a) se aprende o aproxima a través de algoritmos iterativos como **Q-Learning** o **SARSA**, los cuales se basan en premios observados y el resultado de transiciones de estado **S** durante las interacciones de los agentes con el ambiente. Los **valores Q** son actualizados con métodos numéricos como la ecuación de Bellman que combina el premio inmediato de la acción actual y el valor estimado de los **Q-valores** del siguiente par estado y acción (s_t+1,a_t+1). El proceso de aprendizaje puede incluir exploración y explotación (altamente recomendable aunque en un porcentaje de exploración o valor de épsilon no tan grande, digamos 0.1), lo que permitirá al agente inteligente ir refinando sus elecciones y los valores Q y, así, converger a la política óptima $\pi^*$.

Cuando los estados y acciones son muchos, se recomiendo utilizar métodos menos costosos en términos computacionales para aproximar los **valores Q** para cada **par estado-acción (s,a) en un ambiente dado** aproximando finalmente a la **función Q(s,a)**, es aquí donde entran las redes neuronales artificiales y dando paso al **Deep Q-Learning**. Estad redes neuronales son conocidas como DQN (*Deep Q Network*).

**La capa de entrada**, recibe los estados en el ambiente específico.

**Las capas ocultas de la red neuronal**, son similares a la vistas previamente: (i) se trata de capas convolucionales, y (ii) seguidas de una función de activación no lineal (podría ser una sinoidal).

**La capa de salida**, produce los valores Q para cada acción que puede ser tomada por el agente inteligente, en cada estado que fue pasado como input en la capa de entrada.

En este sentido, la **capa de salida** de nuestra red neuronal debería tener tres nodos que presentan cada una de las posibles acciones establecidas en el **conjunto de acciones** descrita previamente: {comprar, vender, esperar}.}. Tomando estas acciones como las que corresponden para el nodo, desde el estado dado en la capa inicial o de entrada.


En cuanto al **Double Deep Q-Learning**, presentado por primera vez por Hasselt Hado (2010), propuesto para resolver el problema de las grandes sobreestimaciones de los valores de la acción (**Q-valores**) en el Deep Q-Learning. Las sobre estimaciones se refieren a que el agente inteligente selecciona las acciones no óptimas en cualcuer estado dado soamente porque éstas tienen el máximo Q-valor.

En el **Q-Learning básico**, la política óptima que sigue el agente inteligente es simpre elegir la mejor actión en cualquier estado dado. El supuesto de trás de esta idea es que la mejor acción tiene el máximo **Q-valor** estimado. Sin embargo, el agente inteligente no conoce mucho sobre el ambiente en un inicio, por lo que necesita estimar **Q(s,a)** y actualizar en cada iteración. En este sentido, los Q-valores tienen mucho ruido y no se tiene certeza de que el argmax sea el mejor, generando sesgos positivos en el procedimiento de actualización, haciendo complicado y desordenado el proceso de aprendizaje.

De acuerdo con, este sesgo positivo impacta tremendamente el procedimiento de actualización:

$$\text{Funcion de Perdida} = Q_{best}(s_t, a_t)-Q(s_t, a_t)$$
$$Q_{best}(s_t, a_t) = R_{t+1} + \gamma*maxQ(s_{t+1}, a) \text {# Desconocido}$$

De tal forma que la actualización a partir de la función de Bellman es la siguiente:

$$
Q(s_t, a_t) \leftarrow Q(s_t,a_t) + \alpha(R_{t+1} + \gamma\cdot\max Q(s_{t+1}, a)-Q(s_t,a_t))
$$

**Double Q-Learning Algorithm** usa dos diferentes funciones de acción-valor, $Q$ y $Q'$ como estimadores. Incluso si $Q$ y $Q'$ son ruidosos o tienen sesgo, estos sesgos pueden ser vistos a través de una distribución uniforme y por consigiente ambos tienen el mismo sesgo y, por sonsiguiente, es irrelevante.

El proceso de actualización cambia de la versión de **Q-Learning básico**:

$$\text {Q-Learning Básico}$$
$$
Q(s_t, a_t) \leftarrow Q(s_t,a_t) + \alpha(R_{t+1} + \gamma\cdot\max Q(s_{t+1}, a)-Q(s_t,a_t))
$$

$$\text {Double Q-Learning}$$
$$
Q(s_t, a_t) \leftarrow Q(s_t,a_t) + \alpha(R_{t+1} +\gamma*Q'(s_{t+1},a)-Q(s_t,a_t))
$$

$$
a=maxQ(s_{t+1},a)
$$

$$
q_{estimada}=Q'(s_{t+1}, a)
$$

La función Q es para seleccionar la mejor acción a con máximo **Q-valor** del siguiente estado.

$$
a = maxQ(s_{t+1},a)
$$

$\text{La función } Q'$ calcula el $Q-valor$ esperado usando la acción a seleccionada arriba:

$$
q_{estimada} = Q'(s_{t+1}, a)
$$

Actualiza la función Q usando el valor esperado del $Q-valor$ de la $Q'$ función.

$$
\text{Double Q-Learning}
$$

$$
Q(s_t, a_t) \leftarrow Q(s_t,a_t) + \alpha(R_{t+1} +\gamma*Q'(s_{t+1},a)-Q(s_t,a_t))
$$

El método propuesto genera una matrix donde los renglones son estados y las columnas son las acciones y cada elemento de la matrix es un $Q-valor$ par, especificado como $Q(s,a)$. Es por este motivo que para ahorrar poder de cómputo,**es aconsejable utilizar una red neuronal en lugar de iteración de valores para el aprendizaje**.

### Double Deep Q Network (Double DQN)

En este sentido, el Double Q-Learning que se implementa con una red neuronal es llamado Double Deep Q Network.

Este algoritmo utiliza difrentes Redes Neuronales Profundas:

- Red Neuronal Profunda
- Red Objetivo

En este caso, no se tiene una tasa de aprendizaje $\alpha$ cuando actualizamos los $Q-valores$ desde que estos serán usados en la etapa de optimización de actualización de los parámetros Deep Q Network:

$$
\text{Double Q-Network}
$$

$$
Q_{qnet}(s_t, a_t) \leftarrow R_{t+1}+\gamma*Q_{tnet}(s_{t+1},a)
$$

$$
a = maxQ_{qnet}(s_{t+1},a)
$$

$$
q_{estimada} = Q_{tnet}(s_{t+1}, a)
$$

$$
\text{Red Profunda Q - seleccionando la mejor acción a con el máximo Q-valor del siguiente estado}
$$

$$
a = maxQ_{qnet}(s_{t+1},a)
$$

$$
\text{Red objetivo - calculando el Q-valor estimado con la acción a seleccionada arriba}
$$
$$
q_{estimada} = Q_{tnet}(s_{t+1}, a)
$$

$$
\text{Actualizar el Q-valor de la red profunda Q basado en el Q-valor estimado en la red objetivo}
$$

$$
Q_{qnet}(s_t, a_t) \leftarrow R_{t+1}+\gamma*Q_{tnet}(s_{t+1},a)
$$

$$
\text{Actualizar los parámetros de la red objetivo basados en los parámetros de la red profunda Q por varias iteraciones}
$$

$$
Q_{tnet}(s,a) = Q_{qnet}(s,a)
$$

Posteriormente, se procurará que las acciones de exploración por parte del agente inteligente se reduzca y, en su lugar, utilizar el modelo entrenado, por lo que estabeceremos un $\epsilon=0.1$, para no eliminar la posibilidad de que nuestro robot siga aprendiendo. En este último punto se definirá una función de decaimiento de $\epsilon$.




```python
# con estos parámetros en mente, se define al bot_broker a partir de un método contructor de una clase, con los atributos correspondientes
# En concordancia con la progamación orientada a objetos vista inicialmente:

class Bot_Broker():

  def __init__(self, state_size, action_space=3, model_name="BotBroker_DQL"):
    self.state_size = state_size
    self.action_space = action_space
    self.memory = deque([])
    self.inventory = []
    self.model_name = model_name

    self.gamma = 0.95 # lo que proporciona mucha importancia a las recompensas futuras con respecto al presente (es un agente inteligente muy paciente)
    self.epsilon = 1.0 # Inicia como un agente inteligente que tomará acciones aleatorias en su totalidad (no será 'greedy' en un principio)
    self.epsilon_final = 0.10 # Eventualmente el agente inteligente tendrá una estrategia balanceada entre argmax y random en un 10% de sus acciones.
    self.epsilon_decay = 0.995 # Épsilon decaerá de 1.0 a (1*0.995) en la primera iteración de entrenamiento.

  # Estos hiperparámetros controlan el balance entre exploración y explotación durante el proceso de entrenamiento del agente
  # La función __init__ inicializalos atributos e hiperparámetros de la clase agente, fijando los componentes necesarios para el algoritmo de aprendizaje por refuerzo.

  # Crear y compilar el modelo
    self.model = self.red_neuronal()
    self.model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

  def red_neuronal(self):
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(units=32, activation='relu', input_shape=(state_size,)))
    # model.add(tf.keras.layers.Dense(units=64, activation='relu')) # possible modifications
    # model.add(tf.keras.layers.Dense(units=128, activation='relu')) # possible modifications
    model.add(tf.keras.layers.Dense(units=self.action_space, activation='linear'))
    return model

  def trade(self, state):
    if random.random() <= self.epsilon:
        return random.randrange(self.action_space)

    state = np.reshape(state, (state.shape[0], -1))  # Reshape the state
    actions = self.model.predict(state)
    if actions is None or len(actions) == 0:
        return random.randrange(self.action_space)

    return np.argmax(actions[0])

  def batch_train(self, batch_size, states, actions):
    if len(self.memory) < batch_size:
        return

    batch = []
    for i in range(len(self.memory) - batch_size, len(self.memory)):
        batch.append(self.memory[i])

    padded_states = []
    padded_targets = []

    for sample in batch:
        state, action, reward, next_state, done = sample

        if state is not None:
            state = np.reshape(state, (state.shape[0], -1))  # Reshape the state
            padded_states.append(state)

            target = reward
            if not done:
                next_state = np.reshape(next_state, (next_state.shape[0], -1))  # Reshape the next state
                target = reward + self.gamma * np.amax(self.model.predict(next_state)[0])
            target_f = self.model.predict(state)
            target_f[0][action] = target
            padded_targets.append(target_f)

    if padded_states and padded_targets:
        padded_states = np.vstack(padded_states)
        padded_targets = np.vstack(padded_targets)
        self.model.fit(padded_states, padded_targets, epochs=1, verbose=0)

    self.memory = self.memory

  def sigmoid(self, x, deriv=False):
    if deriv:
        return x * (1 - x)
    return 1 / (1 + np.exp(-x))

  def state_creator(self, data, t, window_size):
    starting_id = t - window_size + 1 if t >= window_size else 0
    if len(data[starting_id: t + 1]) != window_size:
        return None

    state = np.array(data[starting_id: t + 1])
    state = np.pad(state, ((window_size - state.shape[0], 0), (0, 0)), mode='constant')
    return state

  def save_model(self, file_path):
    self.model.save(file_path)
```

# Construcción del bot_broker (Política)

Una vez definida la red neuronal se construye la función de comercialización que tomará el estado como un input y regresará la acción a desempeñar en el estado específico.

La función se llamará $trade$ que toma como argumento el estado: $state$.

Para cada estado, será necesario determinar si deberíamos usar una acción aleatoria generada aleatoriamente o la red neuronal.

Para hacer esto, usaremos la librería **random** y si es menor a épsilon se genera una acción aleatoria con ***random.randrange()** y que ocurre en **self.action_space = 3**.

En este caso se obtendrá un número con np.argmax para una estrategia 'greedy'.

- La función toma como insumo la matriz con las cotizaciones y genera un número aleatorio.

- Su el número es menor o igual a épsilon generará una acción aleatoria en el recorrido {comprar, vender, esperar} (siempre será el caso inicial).

- Si esto es mayor que épsilon, se elegirá el argmax como una estrategia 'greedy'.

```python
  def trade(self, state):
    if random.random() <= self.epsilon:
        return random.randrange(self.action_space)

    actions = self.model.predict(state)
    if actions is None or len(actions) == 0:
        return random.randrange(self.action_space)

    return np.argmax(actions[0])
```

# Función de entrenamiento

Esta función tomará el lote de datos salvados y entrenará el modelo. El proceso de entrenamiento implica los siguientes pasos:
1. Se define la función **batch_trade** y tomará como argumento **batch_size**.
2. Se seleccionan los datos de la memoria de repetición de la experiencia para el primer conjunto de datos **batch** a una lista vacía.
3. Se itera atravéz de la memoria con un loop 'for'.
4. Como se trata de series de tiempo una muestra de el final de la memoria en lugar de una muestra aleatoria.
5. Una vez obtenidos los lotes de datos se requiere iterar a través de cada lote -state, reward, next_state y done- y entrenar el modelo con esta secuencia.
6. Si el agente inteligente no se encuentra en el estado terminal, se calcula el premio total descontado, como el premio actual.
7. Se define la variable objetivo (**target**) que también es pronósticada por el modelo.
8. Después de entrenar el modelo con **self.model.fit()**
9. Al final de esta función se presenta la regla de decaimiento de épsilon.

```python
def batch_train(self, batch_size):
    if len(self.memory) < batch_size:
        return
    batch = random.sample(self.memory, batch_size)

    states = []
    targets = []

    for experience in batch:
        state, action, reward, next_state, done = experience

        if state is None or next_state is None:
            continue

        states.append(state)
        target = reward
        if not done:
            target += self.gamma * np.amax(self.model.predict(next_state)[0])
        targets.append(target)

    if not states or not targets:
        return

    states = np.array(states)
    targets = np.array(targets)

    states = np.squeeze(states)
    targets = np.squeeze(targets)

    self.model.fit(states, targets, epochs=1, verbose=0)

    if self.epsilon > self.epsilon_final:
        self.epsilon *= self.epsilon_decay
```

Una vez que se tiene la función **dataset_loader** necesitamos crear una función que tome estos datos y genere los estados a los que se enfrentará el agente inteligente.

Lo anterior tomando en cuenta lo siguiente:

- Cada punto en la gráfica del precio de la cción es un número con formato floating que representa un precio de la acción a cada punto en el tiempo.

- Nuestra tarea es predecir qué pasará en el siguiente periodo y, en función de ello, el agente inteligente seleccionará una de las tres acciones posibles: comprar, vender o esperar.

- De acuerdo con la literatura, se trata de un problema de regresión, digamos que tenemos una ventana de tamaño 5 (**window_size=5**), en este sentido se utilizarán 5 estados para predecir qué acción es la mejor o más óptima.

- Se cambiarán nuestros insumos en los estados en función de las diferencias en los precios de las acciones, lo que representará cambios a lo largo del tiempo.

Para implementar lo anterior, se crea la función **state_creator**, el cual toma 3 argumentos: **data**, **timesteps** y **window_size=5**.

1. Se calcula el **starting_id**
2. Cuando el **starting_id** es positivo se crea un estado y si este estado estado inicial es negativo agregamos la información hasta que tenemos la **window_size**.
3. Posteriormente definimos una lista vacía llamada **state** e iteramos a travez de la lista **window_data**.
4. En la medida en que agregamos el estado necesitamos normalizar el precio de los datos con la función sigmoide.
5. Para completar la función necesitamos un arreglo numpy del estado.


```python
def state_creator(self, data, timestep, window_size):
    starting_idx = max(timestep - window_size + 1, 0)
    windowed_data = data[starting_idx:timestep + 1]
    normalized_data = [price / windowed_data[0] - 1 for price in windowed_data]

    state = []
    for i in range(window_size - 1):
        if i + 1 < len(normalized_data):
            state.append(self.sigmoid(normalized_data[i + 1] - normalized_data[i]))
    return np.array([state])
```

## Implementación del Modelo Double Deep Q-Learning
Recordar que este algoritmo involucra a dos redes neuronales, la red neuronal principal (red online) y el modelo objetivo.

```python
class Bot_Broker2():
    def __init__(self, state_size, action_space=3, model_name="BotBroker_DDQL"):
        self.state_size = state_size
        self.action_space = action_space
        self.memory = deque([])
        self.inventory = []
        self.model_name = model_name

        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_final = 0.10
        self.epsilon_decay = 0.995

        self.model = self.build_neural_network()
        self.target_model = self.build_neural_network()
        self.target_model.set_weights(self.model.get_weights())

    def build_neural_network(self):
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=32, activation='relu', input_shape=(self.state_size,)))
        # model.add(tf.keras.layers.Dense(units=64, activation='relu')) # possible modifications
        # model.add(tf.keras.layers.Dense(units=128, activation='relu')) # possible modifications
        return model

    def trade(self, state):
        if random.random() <= self.epsilon:
            return random.randrange(self.action_space)

        state = np.reshape(state, (1, self.state_size))
        actions = self.model.predict(state)
        return np.argmax(actions[0])

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return

        batch = random.sample(self.memory, batch_size)

        states = []
        targets = []

        for state, action, reward, next_state, done in batch:
            target = reward
            if not done:
                next_state = np.reshape(next_state, (1, self.state_size))
                target = reward + self.gamma * np.amax(self.target_model.predict(next_state)[0])

            state = np.reshape(state, (1, self.state_size))
            target_f = self.model.predict(state)
            target_f[0][action] = target

            states.append(state)
            targets.append(target_f)

        states = np.vstack(states)
        targets = np.vstack(targets)
        self.model.fit(states, targets, epochs=1, verbose=0)

    def target_train(self):
        self.target_model.set_weights(self.model.get_weights())

    def epsilon_decay_update(self):
        if self.epsilon > self.epsilon_final:
            self.epsilon *= self.epsilon_decay
    
    def batch_train(self, batch_size, states, actions):
        if len(self.memory) < batch_size:
            return

        batch = []
        for i in range(len(self.memory) - batch_size, len(self.memory)):
            batch.append(self.memory[i])

        padded_states = []
        padded_targets = []

        for sample in batch:
            state, action, reward, next_state, done = sample
            
            if state is not None:
                state = np.reshape(state, (state.shape[0], -1))  # Reshape the state
                padded_states.append(state)

                target = reward
                if not done:
                    next_state = np.reshape(next_state, (next_state.shape[0], -1))  # Reshape the next state
                    target = reward + self.gamma * np.amax(self.model.predict(next_state)[0])
                target_f = self.model.predict(state)
                target_f[0][action] = target
                padded_targets.append(target_f)

        if padded_states and padded_targets:
            padded_states = np.vstack(padded_states)
            padded_targets = np.vstack(padded_targets)
            self.model.fit(padded_states, padded_targets, epochs=1, verbose=0)

        self.memory = self.memory

    
    def state_creator(self, data, t, window_size):
        starting_id = t - window_size + 1 if t >= window_size else 0
        if len(data[starting_id: t + 1]) != window_size:
            return None

        state = np.array(data[starting_id: t + 1])
        state = np.pad(state, ((window_size - state.shape[0], 0), (0, 0)), mode='constant')
        return state

    def save_model(self, file_path):
        self.model.save(file_path)
```

### Una vez revisada la literatura respecto a las mejores métricas financieras para la toma de decisiones en materia de transacciones en el mercado accionario, se optó por el **Índice de fuerza relativa** (Relative Strength Index, por sus siglas en inglés), en virtud de lo siguiente: 

- Se trata de una **métrica financiera que podría complementar el aprendizaje del agente inteligente respecto al mejor momento de vender o comprar acciones**, recordemos que el objetivo principal de nuestro agente inteligente es: maximizar los beneficios obtenidos de comprar barato y vender caro en un horizonte de tiempo específico que define el entorno en el que tomará sus decisiones.

- Se trata de un oscilador de impulso que mide la velocidad y el cambio de los movimientos de precios. Se utiliza para identificar condiciones de sobrecompra y sobreventa en una acción u otro instrumento financiero. El RSI oscila entre 0 y 100, proporcionando información sobre la fuerza de los movimientos de precios. Se calcula de la siguiente forma.

$$
RSI = 100 - \left( \frac{100}{{1 + RS}} \right)
$$

- RS se refiere a la fortaleza relativa que se obtiene de dividir el promedio ganado por el promedio perdido para obtener la fortaleza relativa. 

- Su interpretación es la siguiente: 

    - RSI>70 implica que las acciones están sobre-compradas, en realidad es una señal de que el precio se va a corregir y, por consiguiente es una señal de que se debe vender.
    - ESI < 30 implica que las acciones están sobre-vendidas, lo que significa que ha experimentado pérdidas significativas y podría estar siendo subvaluada. Una señal potencial de compra.

Es importante mencionar que esta métrica financiera no debe ser la única a considerar y debe ser complementada por más factores de la dinámica de la industria antes de tomar la decisión de invertir con un conocimiento amplio de las tendencias de la empresa. 

> A pesar de lo anterior, **no hay duda de que el índice de fuerza relativa es una pieza de información relevante para nuestro agente inteligente** razón suficiente para incluirla como una variable adicional en el entorno.

> Una regla más general, que se incluirá para enriquecer el entorno del agente inteligente es sostenido por los brokers y la literatura que sugiere **buscar las brechas o divergencias entre el RSI y el precio de cierre**. Si el precio aumenta mientras el RSI baja, se interpreta como que existe una condición de sobre-venta, lo que implicará que el precio se corregirá a la baja **señal de que se debe vender** mientras que si la brecha es a la inversa (precio de cierre a la baja y RSI a la alza), implicará que se debe comprar, toda vez que existe una condición de sobre-compra y el precio se corregirá a la alza.
```python
class Bot_Broker3():
    def __init__(self, state_size, action_space=3, model_name="BotBroker_DDQL"):
        self.state_size = state_size
        self.action_space = action_space
        self.memory = deque([])
        self.inventory = []
        self.model_name = model_name

        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_final = 0.10
        self.epsilon_decay = 0.995

        self.model = self.build_neural_network()
        self.target_model = self.build_neural_network()
        self.target_model.set_weights(self.model.get_weights())

    def build_neural_network(self):
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=32, activation='relu', input_shape=(self.state_size,)))
        # model.add(tf.keras.layers.Dense(units=64, activation='relu')) # possible modifications
        # model.add(tf.keras.layers.Dense(units=128, activation='relu')) # possible modifications
        model.compile(optimizer='adam', loss='mse')  # Add this line
        return model

    def trade(self, state, growth_rate_diff):
        if random.random() <= self.epsilon:
            return random.randrange(self.action_space)

        state = np.reshape(state, (1, self.state_size))
        actions = self.model.predict(state)

        if growth_rate_diff < -0.2:
            return 2  # Reinforce sell action
        elif growth_rate_diff > 0.2:
            return 1  # Reinforce buy action

        return np.argmax(actions[0])

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return

        batch = random.sample(self.memory, batch_size)

        states = []
        targets = []

        for state, action, reward, next_state, done in batch:
            target = reward
            if not done:
                next_state = np.reshape(next_state, (1, self.state_size))
                target = reward + self.gamma * np.amax(self.target_model.predict(next_state)[0])

            state = np.reshape(state, (1, self.state_size))
            target_f = self.model.predict(state)
            target_f[0][action] = target

            states.append(state)
            targets.append(target_f)

        states = np.vstack(states)
        targets = np.vstack(targets)
        self.model.fit(states, targets, epochs=1, verbose=0)

    def target_train(self):
        self.target_model.set_weights(self.model.get_weights())

    def epsilon_decay_update(self):
        if self.epsilon > self.epsilon_final:
            self.epsilon *= self.epsilon_decay
    
    def batch_train(self, batch_size, states, actions):
        if len(self.memory) < batch_size:
            return

        batch = random.sample(self.memory, batch_size)

        padded_states = []
        padded_targets = []

        for sample in batch:
            state, action, reward, next_state, done = sample
        
            if state is not None:
                state = np.reshape(state, (state.shape[0], -1))  # Reshape the state
                padded_states.append(state)

                target = reward
                if not done:
                    next_state = np.reshape(next_state, (next_state.shape[0], -1))  # Reshape the next state
                    target = reward + self.gamma * np.amax(self.target_model.predict(next_state)[0])
                target_f = self.model.predict(state)
                target_f[0][action] = target
                padded_targets.append(target_f)

        if padded_states and padded_targets:
            padded_states = np.vstack(padded_states)
            padded_targets = np.vstack(padded_targets)
            self.model.fit(padded_states, padded_targets, epochs=1, verbose=0)

    # Clear the memory after training
        self.memory = deque()

    def state_creator(self, data, t, window_size):
        starting_id = t - window_size + 1 if t >= window_size else 0
        if len(data[starting_id: t + 1]) != window_size:
            return None

        state = np.array(data[starting_id: t + 1])
        state = np.pad(state, ((window_size - state.shape[0], 0), (0, 0)), mode='constant')
        return state

    def save_model(self, file_path):
        self.model.save(file_path)
```

# Double Deep Q-Learning con modificación del entorno con un nuevo feature: Diferencias entre las tasas de crecimiento de **Índice de Fuerza Relativa y Precios de Cierre** (state_size=2) con una red neuronal más profunda que permita considerar más features para el entrenamiento y prueba del modelo.

```python
class Bot_Broker4():
    def __init__(self, state_size, action_space=3, model_name="BotBroker_DDQL_complex"):
        self.state_size = state_size
        self.action_space = action_space
        self.memory = deque([])
        self.inventory = []
        self.model_name = model_name

        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_final = 0.10
        self.epsilon_decay = 0.995

        self.model = self.build_neural_network()
        self.target_model = self.build_neural_network()
        self.target_model.set_weights(self.model.get_weights())

    def build_neural_network(self):
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=32, activation='relu', input_shape=(self.state_size,)))
        model.add(tf.keras.layers.Dense(units=64, activation='relu')) # se agrega una capa oculta adicional con 64 neuronas
        model.add(tf.keras.layers.Dense(units=128, activation='relu')) # se agrega una segunda capa oculta adicional con 128 neuronas
        model.compile(optimizer='adam', loss='mse')  # compilación del modelo
        return model

    def trade(self, state, growth_rate_diff):
        if random.random() <= self.epsilon:
            return random.randrange(self.action_space)

        state = np.reshape(state, (1, self.state_size))
        actions = self.model.predict(state)

        if growth_rate_diff < -0.2:
            return 2  # Reforzando la acción de venta
        elif growth_rate_diff > 0.2:
            return 1  # Reforzando la acción de compra

        return np.argmax(actions[0])

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return

        batch = random.sample(self.memory, batch_size)

        states = []
        targets = []

        for state, action, reward, next_state, done in batch:
            target = reward
            if not done:
                next_state = np.reshape(next_state, (1, self.state_size))
                target = reward + self.gamma * np.amax(self.target_model.predict(next_state)[0])

            state = np.reshape(state, (1, self.state_size))
            target_f = self.model.predict(state)
            target_f[0][action] = target

            states.append(state)
            targets.append(target_f)

        states = np.vstack(states)
        targets = np.vstack(targets)
        self.model.fit(states, targets, epochs=1, verbose=0)

    def target_train(self):
        self.target_model.set_weights(self.model.get_weights())

    def epsilon_decay_update(self):
        if self.epsilon > self.epsilon_final:
            self.epsilon *= self.epsilon_decay
    
    def batch_train(self, batch_size, states, actions):
        if len(self.memory) < batch_size:
            return

        batch = random.sample(self.memory, batch_size)

        padded_states = []
        padded_targets = []

        for sample in batch:
            state, action, reward, next_state, done = sample
        
            if state is not None:
                state = np.reshape(state, (state.shape[0], -1))  # cambiando la forma del estado
                padded_states.append(state)

                target = reward
                if not done:
                    next_state = np.reshape(next_state, (next_state.shape[0], -1))  # Cambiando la forma del siguiente estado
                    target = reward + self.gamma * np.amax(self.target_model.predict(next_state)[0])
                target_f = self.model.predict(state)
                target_f[0][action] = target
                padded_targets.append(target_f)

        if padded_states and padded_targets:
            padded_states = np.vstack(padded_states)
            padded_targets = np.vstack(padded_targets)
            self.model.fit(padded_states, padded_targets, epochs=1, verbose=0)

    # Limpiar la memoria después del entrenamiento
        self.memory = deque()

    def state_creator(self, data, t, window_size):
        starting_id = t - window_size + 1 if t >= window_size else 0
        if len(data[starting_id: t + 1]) != window_size:
            return None

        state = np.array(data[starting_id: t + 1])
        state = np.pad(state, ((window_size - state.shape[0], 0), (0, 0)), mode='constant')
        return state

    def save_model(self, file_path):
        self.model.save(file_path)
```
